```python
l = [[1,'a',['cat'],2],[[[3]],'dog'],4,5]
l2 = []
def flat_list(n):
    for i in n:
        if isinstance(i, list):
            flat_list(i)
        else:
            l2.append(i)

flat_list(l)
print(l2)
```
